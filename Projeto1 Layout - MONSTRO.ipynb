## Pará vs Distrito Federal

___
## O projeto irá analisar e comparar as tecnologias na telecomunicação em casas de alvenaria no estado do Pará e Distrito federal, sendo eles respectivamente, um dos menores e o maior IDH do país. Deste modo, será possível ver a discrepancia entre os dois.


___
## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

Criação Dara Frame para 2015:

```python
dados2015 = pd.read_csv('DOM2015.csv',sep=',')
df = pd.DataFrame({"TV_por_Assinatura":dados2015["V02272"],
                   "TV_Fina":dados2015["V02270"],
                   "Acesso_a_internet_em_casa":dados2015["V02322"],
                   "Usa_Celular_para_Internet":dados2015["V02324"],
                   "Tem_Smart_TV":dados2015["V02326"],
                   "Estado": dados2015["UF"],
                   "Material_construção": dados2015["V0203"],
                   "Tem_Tablet": dados2015["V02321"],
                   "Tem_Microcomputador": dados2015["V0231"]})
```

Colocando variáveis em "category" e colocando filtros:

```python
sloq9 = df.TV_por_Assinatura.astype('category')
sloq9.cat.categories = (["Sim", "Não"])
slow9 = df.Acesso_a_internet_em_casa.astype('category')
slow9.cat.categories = (["Sim", "Não"])
sloe9 = df.Tem_Smart_TV.astype('category')
sloe9.cat.categories = (["Sim", "Não"])
slor9 = df.Usa_Celular_para_Internet.astype('category')
slor9.cat.categories = (["Sim", "Não"])
slot9 = df.Tem_Tablet.astype('category')
slot9.cat.categories = (["Sim", "Não"])
sloy9 = df.TV_Fina.astype('category')
sloy9.cat.categories = (["Nenhuma", "Uma", "Duas", "Três", "Quatro", "Cinco", "Seis", "Sete", "Oito", "Nove"])
slou9 = df.Tem_Microcomputador.astype('category')
slou9.cat.categories = (["Sim", "Não"])

dataframe = pd.DataFrame({"TV_por_Assinatura":sloq9,
                   "TV_Fina":sloy9,
                   "Acesso_a_internet_em_casa":slow9,
                   "Usa_Celular_para_Internet":slor9,
                   "Tem_Smart_TV":sloe9,
                   "Estado": dados2015["UF"],
                   "Material_construção": dados2015["V0203"],
                   "Tem_Tablet": slot9,
                   "Tem_Microcomputador":slou9})

a9 = dataframe[dataframe.Estado==15]
b9 = dataframe[dataframe.Estado==53]

c9 = a9[a9.Material_construção == 1]
d9 = b9[b9.Material_construção == 1]
```

Tirando e modificando os Nan:

```python
c9[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]] = c9[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]].fillna(value="Não")
d9[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]] = d9[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]].fillna(value="Não")

dadosA9 = c9.dropna()
dadosDF9= d9.dropna()
```

Salvando o Data Frame:

```python
ParaSalvo9 = dadosA9.to_csv("Para.csv")
DFSalvo9 = dadosDF9.to_csv("DistritoFederal.csv")
```

```python
ut19 = dadosA9.TV_Fina.value_counts(sort=False)
ut29 = dadosA9.TV_por_Assinatura.value_counts(sort=False)
ut39 = dadosA9.Acesso_a_internet_em_casa.value_counts(sort=False)
ut49 = dadosA9.Usa_Celular_para_Internet.value_counts(sort=False)
ut59 = dadosA9.Tem_Smart_TV.value_counts(sort=False)
ut69 = dadosA9.Tem_Tablet.value_counts(sort=False)
ut79 = dadosA9.Tem_Microcomputador.value_counts(sort=False)


DF19 = dadosDF9.TV_Fina.value_counts(sort=False)
DF29 = dadosDF9.TV_por_Assinatura.value_counts(sort=False)
DF39 = dadosDF9.Acesso_a_internet_em_casa.value_counts(sort=False)
DF49 = dadosDF9.Usa_Celular_para_Internet.value_counts(sort=False)
DF59 = dadosDF9.Tem_Smart_TV.value_counts(sort=False)
DF69 = dadosDF9.Tem_Tablet.value_counts(sort=False)
DF79 = dadosDF9.Tem_Microcomputador.value_counts(sort=False)
```

Plotando os Gráficos:

```python
plot9 = ut19.plot(kind='bar',title='TV Fina no Pará 2015',figsize=(6, 6),color=('b'))
plt.ylabel('Quantidade pessoas')
plt.xlabel('números de Televisões')
plt.show()

plot = DF19.plot(kind='bar',title='TV Fina no DF 2015',figsize=(6, 6),color=('b'))
plt.ylabel('Quantidade pessoas')
plt.xlabel('números de Televisões')
plt.show()
```
