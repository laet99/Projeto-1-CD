## Pará vs Distrito Federal

___
## O projeto irá analisar e comparar as tecnologias na telecomunicação em casas de alvenaria no estado do Pará e Distrito federal, sendo eles respectivamente, um dos menores e o maior IDH do país. Deste modo, será possível ver a discrepancia entre os dois.


___
## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_csv('DOM2013.csv',sep=',')
df = pd.DataFrame({"TV_por_Assinatura":dados["V02272"],
                   "TV_Fina":dados["V02270"],
                   "Acesso_a_internet_em_casa":dados["V02322"],
                   "Usa_Celular_para_Internet":dados["V02324"],
                   "Tem_Smart_TV":dados["V02326"],
                   "Estado": dados["UF"],
                   "Material_construção": dados["V0203"],
                   "Tem_Tablet": dados["V02321"],
                   "Tem_Microcomputador": dados["V0231"]})

```

```python
sloq = df.TV_por_Assinatura.astype('category')
sloq.cat.categories = (["Sim", "Não"])
slow = df.Acesso_a_internet_em_casa.astype('category')
slow.cat.categories = (["Sim", "Não"])
sloe = df.Tem_Smart_TV.astype('category')
sloe.cat.categories = (["Sim", "Não"])
slor = df.Usa_Celular_para_Internet.astype('category')
slor.cat.categories = (["Sim", "Não"])
slot = df.Tem_Tablet.astype('category')
slot.cat.categories = (["Sim", "Não"])
sloy = df.TV_Fina.astype('category')
sloy.cat.categories = (["Nenhuma", "Uma", "Duas", "Três", "Quatro", "Cinco", "Seis", "Sete", "Oito", "Nove"])
slou = df.Tem_Microcomputador.astype('category')
slou.cat.categories = (["Sim", "Não"])

dataframe = pd.DataFrame({"TV_por_Assinatura":sloq,
                   "TV_Fina":sloy,
                   "Acesso_a_internet_em_casa":slow,
                   "Usa_Celular_para_Internet":slor,
                   "Tem_Smart_TV":sloe,
                   "Estado": dados["UF"],
                   "Material_construção": dados["V0203"],
                   "Tem_Tablet": slot,
                   "Tem_Microcomputador":slou})

a = dataframe[dataframe.Estado==15]
b = dataframe[dataframe.Estado==53]

c = a[a.Material_construção == 1]
d = b[b.Material_construção == 1]
```

```python
c[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]] = c[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]].fillna(value="Não")
d[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]] = d[['Acesso_a_internet_em_casa', 'Tem_Smart_TV', "TV_por_Assinatura", "Usa_Celular_para_Internet"]].fillna(value="Não")
```

```python
#Filtro Nan
dadosA = c.dropna()
dadosDF= d.dropna()
```

### salvando
AmapaSalvo = dadosA.to_csv("Amapá.csv")
DFSalvo = dadosDF.to_csv("DistritoFederal.csv")

```python
ut1 = dadosA.TV_Fina.value_counts(sort=False)
ut2 = dadosA.TV_por_Assinatura.value_counts(sort=False)
ut3 = dadosA.Acesso_a_internet_em_casa.value_counts(sort=False)
ut4 = dadosA.Usa_Celular_para_Internet.value_counts(sort=False)
ut5 = dadosA.Tem_Smart_TV.value_counts(sort=False)
ut6 = dadosA.Tem_Tablet.value_counts(sort=False)
ut7 = dadosA.Tem_Microcomputador.value_counts(sort=False)


DF1 = dadosDF.TV_Fina.value_counts(sort=False)
DF2 = dadosDF.TV_por_Assinatura.value_counts(sort=False)
DF3 = dadosDF.Acesso_a_internet_em_casa.value_counts(sort=False)
DF4 = dadosDF.Usa_Celular_para_Internet.value_counts(sort=False)
DF5 = dadosDF.Tem_Smart_TV.value_counts(sort=False)
DF6 = dadosDF.Tem_Tablet.value_counts(sort=False)
DF7 = dadosDF.Tem_Microcomputador.value_counts(sort=False)
```

```python
plot = ut1.plot(kind='bar',title='TV Fina no Pará',figsize=(6, 6),color=('b'))
plt.ylabel('Quantidade pessoas')
plt.xlabel('números de Televisões')
plt.show()

plot = DF1.plot(kind='bar',title='TV Fina no DF',figsize=(6, 6),color=('b'))
plt.ylabel('Quantidade pessoas')
plt.xlabel('números de Televisões')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 5))

plt.subplot(131)
plot = ut2.plot(kind='pie',title='TV por Assinatura (Pará)',autopct='%.1f')
plt.subplot(132)
plot = DF2.plot(kind='pie',title='TV por Assinatura (DF)',autopct='%.1f')

plt.tight_layout()
plt.show()
```
